# Sqwale 🐋
This notebook installs Rust, clones the Sqwale repository, and tests the CUDA/TensorRT execution providers using the free Google Colab T4 GPU.

```python
# 1. Verify we have a GPU attached
!nvidia-smi
```

```python
# 2. Install Rust toolchain
!curl --proto '=https' --tlsv1.2 -sSf https://sh.rustup.rs | sh -s -- -y

# Add cargo to the Python environment path so subsequent cells can use it
import os
os.environ['PATH'] += ":/root/.cargo/bin"
```

```python
# 3. Install FFmpeg (Colab usually has it, but this ensures we have a recent version)
!apt-get update && apt-get install -y ffmpeg
```

```python
# 4. Clone Sqwale and build in release mode
!git clone https://github.com/Hyphonical/Sqwale.git
%cd Sqwale
!cargo build --release
```

```python
# 5. Download some test media
!wget -O test_video.mp4 https://download.blender.org/peach/bigbuckbunny_movies/BigBuckBunny_320x180.mp4
!wget -O test_image.png https://upload.wikimedia.org/wikipedia/commons/thumb/a/a7/React-icon.svg/512px-React-icon.svg.png
```

```python
# 6. Benchmark Image Upscaling with CUDA
# Note: Colab automatically has CUDA installed globally, so ONNX Runtime will find it.
!./target/release/sqwale upscale test_image.png --provider cuda
```

```python
# 7. Benchmark Video Interpolation with CUDA
!./target/release/sqwale interpolate test_video.mp4 --provider cuda
```
